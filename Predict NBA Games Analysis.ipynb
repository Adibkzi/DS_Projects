```python
import pandas as pd 
```

```python
# read the file

nba_df = pd.read_csv("nba_games.csv")

nba_df

```

### Data Cleaning and Data Preparing

```python
nba_df = nba_df.sort_values("date")
nba_df = nba_df.reset_index(drop=True)
```

```python
del nba_df["mp.1"]
del nba_df["mp_opp.1"]
del nba_df["index_opp"]
del nba_df["Unnamed: 0"]
```

```python
def add_target(group):
    group["target"] = group["won"].shift(-1)
    return group

nba_df = nba_df.groupby("team", group_keys=False).apply(add_target)
```

```python
# lets take a look at some stats 
# washington nba_df[nba_df["team"] == "WAS"]

# dallas mavs 
nba_df[nba_df["team"] == 'DAL']
```

```python
# see null values 
nba_df["target"][pd.isnull(nba_df["target"])] = 2
nba_df["target"] = nba_df["target"].astype(int, errors="ignore")
```

```python
# see won there will always be a winner and loser 
nba_df["won"].value_counts()
```

```python
# see the target as well to determine 
nba_df["target"].value_counts()
# 1 win ; 0 = loss 2 = doesnt have next game 
```

```python
nulls = pd.isnull(nba_df).sum()
nulls = nulls[nulls > 0]
valid_columns = nba_df.columns[~nba_df.columns.isin(nulls.index)]
valid_columns
```

```python
# dont want copy warning 
nba_df = nba_df[valid_columns].copy()
nba_df
```

### Best Feature model for Machine Learning 


```python
from sklearn.linear_model import RidgeClassifier
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit

rr = RidgeClassifier(alpha=1)

split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, 
                                n_features_to_select=30, 
                                direction="forward",
                                cv=split,
                                n_jobs=1
                               )
```

```python
# Min and Max Scale to determine 
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
nba_df[selected_columns] = scaler.fit_transform(nba_df[selected_columns])
```
